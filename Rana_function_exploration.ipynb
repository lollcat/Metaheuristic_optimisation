```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 300
from rana import rana_func
#%matplotlib notebook
```

```python
import itertools
n = 1000
x_linspace = np.linspace(-500, 500, n)
x = np.array(list(itertools.product(x_linspace, repeat=2))) 
objectives = np.apply_along_axis(func1d=rana_func, arr=x , axis=1)
x1 = x[:, 0].reshape(n, n)
x2 = x[:, 1].reshape(n, n)
z = objectives.reshape(n, n)
```

```python
from mpl_toolkits.mplot3d import axes3d
offset= -2000
fig = plt.figure()
ax = fig.gca(projection='3d')
trisurf = ax.plot_trisurf(x1.flatten(), x2.flatten(), z.flatten(), cmap=mpl.cm.jet)
cs = ax.contour(x1, x2, z, offset=offset, cmap=mpl.cm.jet, stride = 1, linewidths=0.5)

#fig.subplots_adjust(right=0.8)
#cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.5])
#CB = fig.colorbar(trisurf, label="f(x)", cax=cbar_ax)
#CB.set_ticks([-500, 0, 500])

ax.set_xlabel("x1")
ax.set_ylabel("x2")
ax.set_zlabel("f(x)")
ax.set_xlim(-500, 500)
ax.set_ylim(-500, 500)
ax.set_zlim(offset, 500)
ax.set_zticks([-600, 0, 600])
ax.set_xticks([-500, 0 , 500])
ax.set_yticks([-500, 0, 500])
ax.set_xlabel(r"$x_1$")
ax.set_ylabel(r"$x_2$")
fig.tight_layout()
fig.savefig("./Figures/rana_2D.png")
```
