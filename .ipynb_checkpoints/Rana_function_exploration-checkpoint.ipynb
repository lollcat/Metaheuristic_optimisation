```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 300
#%matplotlib notebook
```

```python
def rana_func(x):
    return np.sum(x[:-1]*np.cos(np.sqrt(np.abs(x[1:]+x[:-1]+1)))*np.sin(np.sqrt(np.abs(x[1:]-x[:-1]+1))) + \
           (1+x[1:])*np.cos(np.sqrt(np.abs(x[1:]-x[:-1]+1)))*np.sin(np.sqrt(np.abs(x[1:]+x[:-1]+1))))

"""
def rana_func(x):
    return np.sum([x[i]*np.cos(np.sqrt(np.abs(x[i+1]+x[i]+1)))*np.sin(np.sqrt(np.abs(x[i+1]-x[i]+1))) + \
           (1+x[i+1])*np.cos(np.sqrt(np.abs(x[i+1]-x[i]+1)))*np.sin(np.sqrt(np.abs(x[i+1]+x[i]+1))) 
                   for i in range(0, len(x)-1)])
"""
```

```python
rana_func(np.array([2,3]))
```

```python
n = 300
x = np.linspace(-500, 500, n)
y = np.linspace(-500, 500, n)
```

```python
z = np.zeros((n,n))
for i, x_val in enumerate(x):
    for j, y_val in enumerate(y):
        z[i,j] = rana_func(np.array([x_val, y_val]))
```

```python
h = plt.contourf(x,y,z)
```

```python
y_grid, x_grid = np.meshgrid(x,y)
```

```python
z.max(), z.min()
```

```python
from mpl_toolkits.mplot3d import axes3d
offset= -1000
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_trisurf(y_grid.flatten(), x_grid.flatten(), z.flatten(), cmap=mpl.cm.jet)
cset = ax.contour(x_grid, y_grid, z, offset=offset, cmap=mpl.cm.jet, stride = 1, linewidths=0.5)
ax.set_xlabel("variable 1")
ax.set_ylabel("variable 2")
ax.set_zlabel("cost")
ax.set_xlim(-500, 500)
ax.set_ylim(-500, 500)
ax.set_zlim(offset, 500)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(y_grid, x_grid, z, cmap=mpl.cm.jet)
ax.set_xlabel("variable 1")
ax.set_ylabel("variable 2")
ax.set_zlabel("cost")
```
